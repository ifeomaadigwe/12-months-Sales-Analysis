# Clean up the data!

### Drop rows with NAN 

```python
nan_data = data[data.isna().any(axis=1)]
nan_data
```

```python
data = data.dropna(how = 'all')
data
```

### Find "OR" within the dataset and delete or correct it

```python
or_data = data[data['Order Date'].str[0:2] == 'Or'] # this shows all the 'Or' in the dataset
or_data
```

```python
data = data[data['Order Date'].str[0:2] != 'Or']# reassigns the variable data without the 'Or'
data
```

### Convert Columns to the Correct Data type

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered']) # make int
data['Price Each'] = pd.to_numeric(data['Price Each']) # make float

data.head()
```

## Add a Month Column

```python
# Create a Month Column

data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int32') # changes the data type of Month to int
data.head()
```

```python
# Replace the month numbers with month names

data['MonthName'] = data['Month'].map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
                                   7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})

data.head()
```

### Abstrate the City Names from the Purchase Address

```python
# we can use .apply() ---- 1

data['City'] = data['Purchase Address'].apply(lambda x : x.split(',')[1])
data.head()
```

```python
# we can use the def method() ----- 2

def get_city(address):
    return address.split(',')[1]
data['City'] = data['Purchase Address'].apply(lambda x : get_city(x))
data.head()
```

### Add a state column to identify the state names from the Purchase Address in the dataframe

```python
data['State'] = data['Purchase Address'].apply(lambda x: x.split(',')[2].split(' ')[1])
data.head()
```

## Add a Sales Column

```python
data['Sales'] = data['Quantity Ordered'] * data['Price Each']
data.head()
```

# Question 4: What products are most often sold together?

### From the dataset, you will notice that some products were bought on the same day and delivered to same address

### - Remove duplicated Order Id

```python
df = data[data['Order ID'].duplicated(keep=False)]
df
```

```python
print(data.columns)
```

```python
df = data[data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df.head()
```

```python
df = data[data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(10)

# Group all the products bought along side with another product
```

```python
## Referenced: https://stackflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary


# To count the number of sales for each grouped in descending order

from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(15):
    print(key, value)
    
# The iPhone and Lightning Charging Cable are mostly bought together
```

# Analysis:
- The most common product combination sold together is the ('iPhone', 'Lightning Charging Cable') pair, with 1005 occurrences.
- Other popular combinations include 'Google Phone' with either 'USB-C Charging Cable' or 'Wired Headphones', and 
- 'Vareebadd Phone' with 'USB-C Charging Cable' or 'Wired Headphones'.
